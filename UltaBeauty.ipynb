# Feature Engineering

Because each product has multiple reviews, we need to aggregate the data between reviews for each product and fine-tune our data for a model/dashboard.

```python
group = merged_df.groupby('product_id')

# Safe numeric aggregation
sentiment_stats = group['sentiment_score'].agg(['mean','min', 'max'])

# Count reviews
review_counts = group['review_text'].count().rename('review_count')

# Get first price
price_usd = group['price_usd'].first()

# Get first category
first_category = group['primary_category'].first().rename('primary_category')

#get second_category

secondary_category = group['secondary_category'].first().rename('secondary_category')

# Get total feedback count

total_feedback = group['total_feedback_count'].sum()

#Get total negative feedback

total_negative_feedback = group['total_neg_feedback_count'].sum()

#Get total positive feedback

total_positive_feedback = group['total_pos_feedback_count'].sum()

# Get total loves count

total_loves_count = group['loves_count'].sum()

#get boolean columns

limited_edition = group['limited_edition'].first().rename('limited_edition')
new_flag= group['new'].first().rename('new')
online_only = group['online_only'].first().rename('online_only')
out_of_stock = group['out_of_stock'].first().rename('out_of_stock')

```

```python
#concat

product_scores = pd.concat([sentiment_stats, review_counts, price_usd, first_category, secondary_category, total_feedback,total_negative_feedback, total_positive_feedback,
                            total_loves_count, limited_edition, new_flag, online_only, out_of_stock], axis=1)

# Extract one row per product from merged_df
product_info = merged_df.drop_duplicates('product_id')[['product_id', 'product_name','brand_name']]

# Merge it back
product_scores = product_scores.reset_index().merge(product_info, on='product_id', how='left')
```

```python
product_scores.rename(columns = {'mean':'mean_sentiment_score', 'max':'max_sentiment_score', 'min':'min_sentient_score'}, inplace=True)
```

```python
product_scores = product_scores.loc[:, ~product_scores.columns.duplicated()]
```
